3.Apply and explore various plotting functions on Pima Indians Diabetes data set for performing the
following:
d) Normal values
e) Density and contour plots
f) Three-dimensional plotting

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px

# Load Pima Indians Diabetes dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00463/"
file = "diabetes.csv"
diabetes_data = pd.read_csv(url + file)

# Display the first few rows of the dataset
print(diabetes_data.head())

# d) Normal values
plt.figure(figsize=(12, 6))
diabetes_data.boxplot()
plt.title("Normal Values Plot for Pima Indians Diabetes Dataset")
plt.show()

# e) Density and Contour Plots
plt.figure(figsize=(12, 8))
sns.kdeplot(data=diabetes_data, x="Glucose", y="BMI", fill=True, cmap="Blues", levels=15)
plt.title("Density and Contour Plot for Glucose vs BMI")
plt.show()

# f) Three-dimensional plotting
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(diabetes_data['Glucose'], diabetes_data['BMI'], diabetes_data['Age'], c=diabetes_data['Outcome'], cmap='viridis', s=50)
ax.set_xlabel('Glucose')
ax.set_ylabel('BMI')
ax.set_zlabel('Age')
ax.set_title('Three-Dimensional Plot: Glucose, BMI, and Age')
plt.show()

# Alternatively, you can use Plotly for interactive 3D plotting
fig = px.scatter_3d(diabetes_data, x='Glucose', y='BMI', z='Age', color='Outcome', opacity=0.7)
fig.update_layout(title="Interactive 3D Plot: Glucose, BMI, and Age", scene=dict(xaxis_title='Glucose', yaxis_title='BMI', zaxis_title='Age'))
fig.show()
```

```python

#normal values plotted
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('diabetes.csv')

# Histogram for distribution
df.hist(figsize=(12, 8))
plt.show()

# Box plots for understanding range and outliers
df.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False, figsize=(12,8))
plt.show()
```

```python
# Density Plot
df.plot(kind='density', subplots=True, layout=(3,3), sharex=False, figsize=(12, 8))
plt.show()

# Contour Plot with seaborn
sns.jointplot(x='Glucose', y='Insulin', data=df, kind='kde')
plt.show()
```
